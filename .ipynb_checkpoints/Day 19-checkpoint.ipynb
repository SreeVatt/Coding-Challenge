```python
# Enter your code here. Read input from STDIN. Print output to STDOUT
thick=int(input())
width=(thick*2)-1
bar=(thick//2)+1
for i in range(1,width+1,2):
    print (('H'*i).center(width,' '))
for i in range(0,thick+1):
    print(('H'*thick).center(width,' '),end='')
    print(' '*(width+2),end='')
    print(('H'*thick).center(width,' '))
for i in range(bar):
    print(('H'*(thick*5)).center((width*3)+2,' '))
for i in range(0,thick+1):
    print(('H'*thick).center(width,' '),end='')
    print(' '*(width+2),end='')
    print(('H'*thick).center(width,' '))   
for i in range(width,0,-2):
    print(' '*((width*2)+2),end='')
    print (('H'*i).center(width,' '))
```

```python
# Enter your code here. Read input from STDIN. Print output to STDOUT
inp=input()

last=0
arr=[]
for i in range(len(inp)):
    if(inp[i]==' '):
        arr.append(int(inp[last:i]))
        last=i+1
arr.append(int(inp[last:]))
for i in range(1,(arr[0]),2):
    print(('.|.'*i).center(arr[1],'-'))
print('WELCOME'.center(arr[1],'-'))
for i in range((arr[0]),0,-2):
    if(i==arr[0]):
        continue
    print(('.|.'*i).center(arr[1],'-'))
```
