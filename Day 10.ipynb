```python
a=int(input())
b=[]
for i in range(a):
  x=[]
  x,y=map(int,input().split())
  b.append([x,y])

maxi=-1000000
maxval=0
minval=0
mini=100000

for i in b:
  if(i[0]>maxi):
    maxval=i[1]
    maxi=i[0]

  if(i[0]==maxi):
    if (maxval>i[1]):
      maxval=i[1]


for i in b:
  if(i[0]<mini):
    minval=i[1]
    mini=i[0] 

  if(i[0]==mini):
    if (minval>i[1]):
      minval=i[1]
      


print(abs(maxval-minval))

  
```
